```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
X = X[:, 2]
```

### This is the third col in a one-dimensional array

```python
print(X[:5])
```

```python
X = X.reshape((-1,1))
```

```python
print(X[:5])
```

After you reshape to a 2D array with `X = X.reshape((-1,1))`, your data has shape `(n_samples, 1)`.  
Each row is a sample, and there is only one column (the third feature).

- `X[i, 0]` gives you the value of the third feature for the `i`-th sample.
- `X[:5, 0]` gives you the first 5 values as a 1D array.
- `X[:5]` gives you the first 5 rows as 2D arrays (each with one column).

**Example:**


```python
print(X[0, 0])      # First value
print(X[:5, 0])     # First 5 values as a flat array
print(X[:5])        # First 5 rows as 2D array
```



Since there is only one column, the only valid column index is `0`.

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
